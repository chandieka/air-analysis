```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_read = pd.read_csv('../datasets/others/merge-date.csv', parse_dates=['date'])
df_read_per_zipcode = dict(tuple(df_read.groupby('PC4')))
```

```python
df_pc1 = df_read_per_zipcode[5611]
# df_pc1.head()
```

```python
df_pc2 = df_read_per_zipcode[5651]
# df_pc2.head()
```

```python
bxplt = df_pc1.boxplot(column=['pm10', 'pm2.5', 'no2', 'no','so2'], figsize=(15,10), fontsize=12)

bxplt.set_xlabel('$Pollutants$', fontsize = 14)
bxplt.set_ylabel('$Pollutants\'\ Values$', fontsize = 14)
bxplt.set_title('Pollutants Distribution', fontsize = 16)
```

```python
bxplt = df_pc2.boxplot(column=['pm10', 'pm2.5', 'no2', 'no','so2'], figsize=(15,10), fontsize=12)

bxplt.set_xlabel('$Pollutants$', fontsize = 14)
bxplt.set_ylabel('$Pollutants\'\ Values$', fontsize = 14)
bxplt.set_title('Pollutants Distribution', fontsize = 16)
```

```python
x = df_pc1.pm10
plt.ylabel('Occurence')
plt.xlabel('pm10')
plt.title("PM10 Occurence Distribution")

plt.hist(x, bins=70)
plt.show()
```

```python
df_specific = df_pc1[(df_pc1['pm10'] >= 56)]
df_specific
```

```python
x = df_pc1['pm2.5']
plt.ylabel('Occurence')
plt.xlabel('pm2.5')
plt.title("PM2.5 Occurence Distribution")

plt.hist(x, bins=50)
plt.show()
```

```python
x = df_pc1.no
plt.ylabel('Occurence')
plt.xlabel('no')
plt.title("NO Occurence Distribution")

plt.hist(x, bins=70)
plt.show()
```

```python
x = df_pc1.no2
plt.ylabel('Occurence')
plt.xlabel('no2')
plt.title("NO2 Occurence Distribution")

plt.hist(x, bins=70)
plt.show()
```

```python
x = df_pc1.so2
plt.ylabel('Occurence')
plt.xlabel('so2')
plt.title("SO2 Occurence Distribution")

plt.hist(x, bins=70)
plt.show()
```

```python
x = [df_pc1['pm10'], df_pc2['pm10']]
plt.ylabel('Occurence')
plt.xlabel('pm10 µg/m3')
plt.title("PM10 Occurence Distribution")

#Add microgram (units) in x-axis
#Conclude (toxic numbers in or not)

plt.hist(x, bins=70, label=['PC 5611 - City Center','PC 5651 - Industrial Area'])
plt.gcf().set_size_inches(20, 10)
plt.legend(prop={'size': 10})
plt.grid(axis = 'y',color = 'lightgrey', linewidth = 0.5)
plt.show()
```

This diagram shows the comparison of the pm10 pollutant in a the city center and the industrial area. This is determine a comparison as well as if there are any outliers within the data. According to this data, there aren't major outliers. Therefore, it can be concluded that all the data for these 2 postal codes are relavant in use of training the model.
